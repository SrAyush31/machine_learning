```python
X = pd.DataFrame(dataset.loc[:,['MonthlyCharges','tenure']])
y = dataset.iloc[:,-1]
```

# **Building Logistic Regression model**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.20)
```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
lm.fit(X_train , y_train)
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred , y_test)
print('Accuracy_score of the Logistic model is:',acc*100,'%')
```

```python
from sklearn.metrics import classification_report
clfr = classification_report(y_test,y_pred)
clfr
```

# **Building Random Forest clf**

```python
X = pd.DataFrame(dataset.loc[:,['MonthlyCharges','tenure']])
y = dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size = 0.20)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=300)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
acc1 = accuracy_score(y_test,y_pred)
print('Accuracy of the random forest clf:',acc1*100,'%')
```
